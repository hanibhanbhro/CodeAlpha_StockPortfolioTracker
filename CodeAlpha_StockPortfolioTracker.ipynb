Stock Portfolio Tracker – How to Use
🎯 Purpose:

This program helps you calculate your total investment in different stocks.

🟢 How to Use:
Run the program.
Enter the stock name when asked:
Allowed stocks: AAPL, TSLA, GOOG, MSFT
Enter the quantity (ONLY numbers like 1, 2, 3).
Keep adding stocks one by one.
When finished, type:
done
The program will show your total investment and summary.
⚠ Important Rules:
Stock names must be written exactly as shown (AAPL, TSLA, etc.)
Quantity must be a number only (not text)
If stock is not in the list, it will show “Stock not found!”
💡 Example:

Enter stock name: AAPL
Enter quantity: 2

Enter stock name: MSFT
Enter quantity: 1

Enter stock name: done

📌 Output:

Program will display:

Each stock total value
Final total investment amount


```python
stocks = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 140,
    "MSFT": 300
}

portfolio = {}

while True:
    stock = input("Enter stock name (or 'done'): ").upper()

    if stock == "DONE":
        break

    if stock in stocks:
        qty = int(input("Enter quantity: "))
        portfolio[stock] = qty
    else:
        print("Stock not found!")

total = 0

print("\n📊 Portfolio Summary:")
for stock, qty in portfolio.items():
    value = stocks[stock] * qty
    print(stock, "x", qty, "=", value)
    total += value

print("\n💰 Total Investment:", total)
```
